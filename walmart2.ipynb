```python
string="i want a laptop with 8gb ram 512gb ssd with aluminum body with solid looks which is light and fast performing with good performance and it should have intel i5 processor,2-in-1 modern design"
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def return_tokens(string):
        
    # Download the stopwords corpus if not already downloaded
    nltk.download('stopwords')

    # Download the punkt tokenizer if not already downloaded
    nltk.download('punkt')

    # Text to process
    text = string
    # Convert the text to lowercase
    text = text.lower()

    # Tokenize the text into individual words
    words = word_tokenize(text)

    # Extract the English stopwords
    english_stopwords = set(stopwords.words('english'))

    # Remove punctuation marks and stopwords
    filtered_words = [word for word in words if word.isalnum() and word not in english_stopwords]

    # Join the filtered words back into a single string
    filtered_text = ' '.join(filtered_words)

    # Create the stopwords list in the desired format
    stopwords_list = [word for word in filtered_words if word.isalnum()]

    # Format the stopwords list as a string
    stopwords_string = str(stopwords_list)

    # Print the stopwords in the desired format
    return stopwords_string

```

```python
def product_similarity(user_preferences, recommended_products):
    # Convert preferences to lowercase
    user_preferences = [pref.lower() for pref in user_preferences]

    recommended_products = [prod.lower() for prod in recommended_products]

    user_pref_len = len(user_preferences)
    recommended_products_len = len(recommended_products)

    matched_products = []
    missing_products = []

    user_pref_count = 0
    prod_count = 0

    for product in recommended_products:
        if product in user_preferences:
            prod_count += 1
            matched_products.append(product)
        else:
            missing_products.append(product)

    # Calculate matching percentage based on user preferences and recommended products
    matching_percentage = (prod_count / user_pref_len) * 100

    print("Matched products: ", matched_products)
    print("Missing products: ", missing_products)
    print("Matching percentage: ", matching_percentage)
    print("Missing percentage: ", (len(missing_products) / user_pref_len) * 100)
    
    if set(recommended_products).issubset(set(user_preferences)):
        interest_similarity = 100
    else:
        # Check the percentage of similarity based on unique preferences
        matching_prefs = set(recommended_products).intersection(set(user_preferences))
        interest_similarity = (len(matching_prefs) / recommended_products_len) * 100

    print("User preferences similarity with recommended products: ", interest_similarity)

    return matching_percentage

```

```python
prompt = return_tokens(string)
string = prompt.replace("[", "").replace("]", "").replace("'", "")
prompt = string.split(", ")

```

```python
import json
import heapq

with open('product.json', 'r') as file:
    products = json.load(file)

# Calculate similarity scores
similarity_scores = {}

for product_name, product_features in products.items():
    product_keywords = []
    for items in product_features:
        product_keywords.extend(items.split())
    similarity_scores[product_name] = len(set(product_features) & set(prompt))

# Find the top 10 products with the highest similarity scores
top_products = heapq.nlargest(10, similarity_scores, key=similarity_scores.get)
print("#############")
print(top_products)
print("#############")
top_products_recommendation = top_products

recommendation_list = []
for product_name in top_products:
    print("Product:", product_name)
    product_keywords = []
    for items in products[product_name]:
        product_keywords.extend(items.split())   
    #print("Features:", product_keywords)
    similarity_percentage = product_similarity(product_keywords, prompt)
    recommendation_list.append({product_name: similarity_percentage})
    print()
print(recommendation_list)
```
